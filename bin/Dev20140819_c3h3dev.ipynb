```python
def get_one_data_column_by_pickle(colname):
    assert colname in list_all_data_columns()
    
    data_col_pickle_path = os.path.join(TRAINING_COLUMNS_PATH,"%s.col") % colname
    
    with open(data_col_pickle_path, "rb") as rf:
        data_col = pickle.load(rf)
        
    return data_col
    
```
