```python
## Modeling.
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
tmp = pd.DataFrame.from_csv("/Users/DboyLiao/Documents/kaggle/data/Display_Advertising_Challenge/train_split_#1.csv")
data = tmp.set_index('Id')
```

```python
data
```

```python
Y = data['Label']
X = data.drop('Label', 1)
```

```python
X.values
```

```python
for i in range(1, 27):
    le = preprocessing.LabelEncoder()
    col = 'C' + str(i)
    X[col].values.astype('a8')
    le.fit(X[col])
    X[col] = le.transform(X[col].values)
X
```

```python
from sklearn.ensemble import RandomForestClassifier as RFC
clf = RFC(n_estimators=20)
fit = clf.fit(X, Y)
```
