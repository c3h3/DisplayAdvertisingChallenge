```python
data_file = open('/Users/DboyLiao/Documents/kaggle/data/Display_Advertising_Challenge/train.csv')
colnames = data_file.readline().strip().split(',')
print colnames
```

```python
lines = data_file.readlines(250000000)
data = [line.strip().split(',') for line in lines]
print data[:3]
```

```python
count = 1
while lines != []:
    file_name = '/Users/DboyLiao/Documents/kaggle/data/Display_Advertising_Challenge/train_split_#' + \
    str(count) + '.csv'
    tmp_df = pd.DataFrame(data, columns = colnames)
    tmp_df.to_csv(file_name, sep = ',', encoding = 'utf-8')
    lines = data_file.readlines(250000000)
    data = [line.strip().split(',') for line in lines]
    count += 1
```

```python
tmp = pd.DataFrame.from_csv("/Users/DboyLiao/Documents/kaggle/data/Display_Advertising_Challenge/train_split_#1.csv")
data = tmp.set_index('Id')
data[:]
```

```python
for i in range(1, 27):
    le = preprocessing.LabelEncoder()
    col = 'C' + str(i)
    data[col].values.astype('a8')
    le.fit(data[col])
    data[col] = le.transform(data[col].values)
data
```

```python
for col in data.columns:
    data = data[np.isfinite(data[col])]
```

```python
file_name = '/Users/DboyLiao/Documents/kaggle/data/Display_Advertising_Challenge/clean_train_split_#1.csv'
data.to_csv(file_name, sep = ',', encoding = 'utf-8')
```
