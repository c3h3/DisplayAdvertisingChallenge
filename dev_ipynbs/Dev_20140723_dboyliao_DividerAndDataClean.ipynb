```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
## Data Slicer.
def data_slicer(file_path, bytes_max):
    data_file = open(file_path)
    colnames = data_file.readline()
    colnames = colnames.strip().split(',')
    lines = [line.strip().split(',') for line in data_file.readlines(bytes_max)]
    count = 1
    file_prefix = file_path[:-4]
    while lines != []:
        file_name = file_prefix + '_split' + '%.4d' % count +'.csv'
        data = pd.DataFrame(lines, columns = colnames)
        data.to_csv(file_name, sep = ',', encoding = 'utf-8')
        lines = [line.strip().split(',') for line in data_file.readlines(bytes_max)]
        count += 1
```

```python
def data_clean(input_path, output_path, C_cols, id_index = False, nan_rm = False):
    print 'Reading ' + input_path
    data = pd.DataFrame.from_csv(input_path)
    if id_index:
        print 'Setting row names into Id.'
        data = data.set_index('Id')
    print "Factorize categorical columns."
    for col in C_cols:
        le = preprocessing.LabelEncoder()
        data[col].values.astype('a8')
        le.fit(data[col])
        data[col] = le.transform(data[col].values)
    if nan_rm:
        print 'Removing NaN observations'
        for col in data.columns:
            data = data[np.isfinite(data[col])]
    print "Output " + output_path
    data.to_csv(output_path)
```

```python
file_path = "/Users/DboyLiao/Documents/kaggle/data/Display_Advertising_Challenge/"
C_cols = ['C' + str(ind) for ind in range(1, 27)]
for count in range(1, 11):
    input_path = file_path + "train_split" + "%.4d" % count + ".csv"
    output_path = file_path + 'clean_train_split' + "%.4d" %count + ".csv"
    data_clean(input_path, output_path, C_cols, id_index = True, nan_rm = True)
```
